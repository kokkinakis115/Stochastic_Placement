```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib as mpl
import seaborn as sns
```

## Model Comparison

```python
df = pd.read_csv("./parameters/results.csv")
```

```python
### Cost

label = "Average Cost"
error_label = "Std Deviation in Cost"

models = ["Fixed Model (1.5 CPUs)", "Fixed Model (2.5 CPUs)",  "B.S. Model (Thr = 10%)", "B.S. Model (Thr = 2%)"]
x_pos = np.arange(len(df[label])*1.5, step=1.5)
value = df[label].to_numpy()
error = df[error_label].to_numpy()

# Build the plot
fig, ax = plt.subplots()
ax.bar(x_pos, value, yerr=error, align='center', alpha=0.5, ecolor='black', capsize=10, color='red')
ax.set_ylabel(label)
plt.xticks(rotation=30, ha='right')
ax.set_xticks(x_pos)
ax.set_xticklabels(models)
ax.set_title(label + " per Prediction Model")
ax.yaxis.grid(True)

# Save the figure and show
plt.tight_layout()
plt.savefig('./plots/cost.png', dpi=1000)
plt.show()
```

```python
### Utilization

label = "Average Resource Utilization Percentage"
error_label = "Std Deviation in Resource Utilization Percentage"

models = ["Fixed Model (1.5 CPUs)", "Fixed Model (2.5 CPUs)",  "B.S. Model (Thr = 10%)", "B.S. Model (Thr = 2%)"]
x_pos = np.arange(len(df[label])*1.5, step=1.5)
value = df[label].to_numpy()*100
error = df[error_label].to_numpy()*100

# Build the plot
fig, ax = plt.subplots()
ax.bar(x_pos, value, yerr=error, align='center', alpha=0.5, ecolor='black', capsize=10, color='orange')
ax.set_ylabel(label + "(%)")
plt.xticks(rotation=30, ha='right')
ax.set_xticks(x_pos)
ax.set_xticklabels(models)
ax.set_title(label + " per Prediction Model")
ax.yaxis.grid(True)

# Save the figure and show
plt.tight_layout()
plt.savefig('./plots/utilization.png', dpi=1000)
plt.show()
```

## Utilization according to Model

```python
# Create objectives Dataframe
df_fixed_15 = pd.read_csv('parameters/usage_fixed_15.csv')
df_fixed_25 = pd.read_csv('parameters/usage_fixed_25.csv')
df_bs_010 = pd.read_csv('parameters/usage_bs_010.csv')
df_bs_002 = pd.read_csv('parameters/usage_bs_02.csv')
```

```python
# Distribution of Utilization According to Objective

fig, ax1 = plt.subplots()

total_usage_fixed_15 = df_fixed_15['Edge'].sum()+df_fixed_15['Fog'].sum()+df_fixed_15['Cloud'].sum()
total_usage_fixed_25 = df_fixed_25['Edge'].sum()+df_fixed_25['Fog'].sum()+df_fixed_25['Cloud'].sum()
total_usage_bs_010 = df_bs_010['Edge'].sum()+df_bs_010['Fog'].sum()+df_bs_010['Cloud'].sum()
total_usage_bs_002 = df_bs_002['Edge'].sum()+df_bs_002['Fog'].sum()+df_bs_002['Cloud'].sum()

edge = np.array([df_fixed_15['Edge'].sum()/total_usage_fixed_15, df_fixed_25['Edge'].sum()/total_usage_fixed_25, df_bs_010['Edge'].sum()/total_usage_bs_010, df_bs_002['Edge'].sum()/total_usage_bs_002])*100
fog = np.array([df_fixed_15['Fog'].sum()/total_usage_fixed_15, df_fixed_25['Fog'].sum()/total_usage_fixed_25, df_bs_010['Fog'].sum()/total_usage_bs_010, df_bs_002['Fog'].sum()/total_usage_bs_002])*100
cloud = np.array([df_fixed_15['Cloud'].sum()/total_usage_fixed_15, df_fixed_25['Cloud'].sum()/total_usage_fixed_25, df_bs_010['Cloud'].sum()/total_usage_bs_010, df_bs_002['Cloud'].sum()/total_usage_bs_002])*100


models = ["Fixed Model (1.5 CPUs)", "Fixed Model (2.5 CPUs)",  "B.S. Model (Thr = 10%)", "B.S. Model (Thr = 2%)"]
x_pos = np.arange(len(models))
ax1.bar(models , edge, width=0.3, align='center', label='Edge')
ax1.bar(models , fog, width=0.3, align='center', label='Fog', bottom=edge)
ax1.bar(models , cloud, width=0.3, align='center', label='Cloud', bottom=[edge[i]+fog[i] for i in range(len(edge))])

# Set labels and title
ax1.set_xlabel('Model')
ax1.set_ylabel('Percentage of CPU Resources Allocated (%)')
plt.xticks(rotation=30, ha='right')

# ax2.set_ylabel('Real Requirements')
plt.title('Distribution of Utilization per Model')

# Display legend
ax1.legend(loc='lower right')

# Show the plot
plt.tight_layout()
# plt.savefig("objectives.png")
plt.show()
```

## Utilization According To Objective

```python
# Create objectives Dataframe
df_latency = pd.read_csv('parameters/usage_latency.csv')
df_cost = pd.read_csv('parameters/usage_cost.csv')
df_balanced = pd.read_csv('parameters/usage_balanced.csv')
```

```python
# Distribution of Utilization According to Objective

fig, ax1 = plt.subplots()

total_usage_cost = df_cost['Edge'].sum()+df_cost['Fog'].sum()+df_cost['Cloud'].sum()
total_usage_balanced = df_balanced['Edge'].sum()+df_balanced['Fog'].sum()+df_balanced['Cloud'].sum()
total_usage_latency = df_latency['Edge'].sum()+df_latency['Fog'].sum()+df_latency['Cloud'].sum()

edge = np.array([df_cost['Edge'].sum()/total_usage_cost, df_balanced['Edge'].sum()/total_usage_balanced, df_latency['Edge'].sum()/total_usage_latency])*100
fog = np.array([df_cost['Fog'].sum()/total_usage_cost, df_balanced['Fog'].sum()/total_usage_balanced, df_latency['Fog'].sum()/total_usage_latency])*100
cloud = np.array([df_cost['Cloud'].sum()/total_usage_cost, df_balanced['Cloud'].sum()/total_usage_balanced, df_latency['Cloud'].sum()/total_usage_latency])*100

objectives = ['Minimize Cost', 'Balanced', 'Minimize Latency']
x_pos = np.arange(len(objectives))
ax1.bar(objectives , edge, width=0.3, align='center', label='Edge')
ax1.bar(objectives , fog, width=0.3, align='center', label='Fog', bottom=edge)
ax1.bar(objectives , cloud, width=0.3, align='center', label='Cloud', bottom=[edge[i]+fog[i] for i in range(len(edge))])

# Set labels and title
ax1.set_xlabel('Objective')
ax1.set_ylabel('Percentage of CPU Resources Allocated (%)')
# ax2.set_ylabel('Real Requirements')
plt.title('Distribution of Utilization per Objective')

# Display legend
ax1.legend(loc='lower right')

# Show the plot
plt.tight_layout()
plt.savefig("plots/objectives.png", dpi=1000)
plt.show()
```
