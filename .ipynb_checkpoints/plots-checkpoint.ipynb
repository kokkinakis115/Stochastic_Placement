```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib as mpl
import seaborn as sns
```

```python
df = pd.read_csv("./parameters/results.csv")
```

```python
label = "Average Cost"
error_label = "Std Deviation in Cost"

models = ["Fixed Model (1.5 CPUs)", "Fixed Model (3.0 CPUs)", "B.S. Model (Thr = 5%)",  "B.S. Model (Thr = 10%)"]
x_pos = np.arange(len(df[label])*1.5, step=1.5)
value = df[label].to_numpy()
error = df[error_label].to_numpy()

# Build the plot
fig, ax = plt.subplots()
ax.bar(x_pos, value, yerr=error, align='center', alpha=0.5, ecolor='black', capsize=10, color='purple')
ax.set_ylabel(label)
plt.xticks(rotation=30, ha='right')
ax.set_xticks(x_pos)
ax.set_xticklabels(models)
ax.set_title(label + " per Prediction Model")
ax.yaxis.grid(True)

# Save the figure and show
plt.tight_layout()
plt.savefig('./plots/cost.png')
plt.show()
```

```python
df = pd.read_csv('usage.csv')
```

```python
time_points = range(145)
# time_points = pd.date_range(start='00:00', periods=145, freq='1d')
# df['dates'] = time_points
# Plotting
fig, ax1 = plt.subplots()

# Plot bars for allocated resources
ax1.bar(time_points , df['Edge'], width=0.3, align='center', label='Edge')
ax1.bar(time_points , df['Fog'], width=0.3, align='center', label='Fog', bottom=df['Edge'])
ax1.bar(time_points , df['Cloud'], width=0.3, align='center', label='Cloud', bottom=df['Edge']+df['Fog'])

# Create another y-axis for real requirements
# ax2 = ax1.twinx()
ax1.plot(time_points, df['Real'], color='r', label='Real Requirements')

# Set labels and title
ax1.set_xlabel('Time (5m)')
ax1.set_ylabel('CPU Resources (Cores)')
# ax2.set_ylabel('Real Requirements')
plt.title('Total Allocated Resources per Layer vs. Real Requirements Over Time')

# Display legend
ax1.legend(loc='lower right')

# Show the plot
plt.tight_layout()
plt.savefig("usage.png")
plt.show()
```
