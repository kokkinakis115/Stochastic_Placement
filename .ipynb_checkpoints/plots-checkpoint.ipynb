```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib as mpl
import seaborn as sns
```

```python
df = pd.read_csv("./parameters/results.csv")
```

```python
label = "Average Cost"
error_label = "Std Deviation in Cost"

models = ["Fixed Model (1.5 CPUs)", "Fixed Model (3.0 CPUs)", "B.S. Model (Thr = 5%)",  "B.S. Model (Thr = 10%)"]
x_pos = np.arange(len(df[label])*1.5, step=1.5)
value = df[label].to_numpy()
error = df[error_label].to_numpy()

# Build the plot
fig, ax = plt.subplots()
ax.bar(x_pos, value, yerr=error, align='center', alpha=0.5, ecolor='black', capsize=10, color='purple')
ax.set_ylabel(label)
plt.xticks(rotation=30, ha='right')
ax.set_xticks(x_pos)
ax.set_xticklabels(models)
ax.set_title(label + " per Prediction Model")
ax.yaxis.grid(True)

# Save the figure and show
plt.tight_layout()
plt.savefig('./plots/cost.png')
plt.show()
```

```python
label = "Average Resource Utilization Percentage"
error_label = "Std Deviation in Resource Utilization Percentage"

models = ["Fixed Model (1.5 CPUs)", "Fixed Model (3.0 CPUs)", "B.S. Model (Thr = 5%)",  "B.S. Model (Thr = 10%)"]
x_pos = np.arange(len(df[label])*1.5, step=1.5)
value = df[label].to_numpy()*100
error = df[error_label].to_numpy()*100

# Build the plot
fig, ax = plt.subplots()
ax.bar(x_pos, value, yerr=error, align='center', alpha=0.5, ecolor='black', capsize=10, color='orange')
ax.set_ylabel(label + "(%)")
plt.xticks(rotation=30, ha='right')
ax.set_xticks(x_pos)
ax.set_xticklabels(models)
ax.set_title(label + " per Prediction Model")
ax.yaxis.grid(True)

# Save the figure and show
plt.tight_layout()
plt.savefig('./plots/utilization.png')
plt.show()
```
